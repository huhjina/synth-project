```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


post_ids = []
post_data = []
unique_id = 0
first_post = ''
forum_url = 'https://www.norduserforum.com/viewforum.php?f=29'
forum_id = forum_url.split('=')[-1]
response = requests.get(forum_url)
soup = BeautifulSoup(response.content, 'html.parser')
views = int(soup.find('ul', class_='topiclist topics').find('dd', class_='views').get_text(strip=True)[:-5])
replies = int(soup.find('ul', class_='topiclist topics').find('dd', class_='posts').get_text(strip=True)[:-7])
thread_url = 'https://www.norduserforum.com/viewtopic.php?t=20656'
thread_id = thread_url.split('=')[-1]
response = requests.get(thread_url)
soup = BeautifulSoup(response.content, 'html.parser')
for div in soup.find_all('div'):
    if div.has_attr('class') and 'post' in div['class']:
        post_ids.append(div['id'])

for post_id in post_ids:
    post_div = soup.find('div', id=post_id)
    username = post_div.find('span', class_='responsive-hide').find('a', href=True)
    if username == None:
        username = post_div.find('span', class_='username')
    username = username.get_text(strip=True)
    datetime = post_div.find('time')['datetime']
    h3tag = post_div.find('h3')
    href = h3tag.find('a', href = True)['href']
    href = href.lstrip('.')
    post_link = f"https://www.norduserforum.com{href}"
    if post_div.find('h3', class_='first'):
        is_it_a_reply = False
        reply_to = ''
        first_post = post_link
    else:
        is_it_a_reply = True
        reply_to = first_post
    if is_it_a_reply:
        title = f"RE: {soup.find('h2', class_='topic-title').get_text(strip=True)}"
    else:
        title = soup.find('h2', class_='topic-title').get_text(strip=True)
    #unique_user_id = post_div.find('div', class_ = 'avatar-container')
    post_data.append({
        'unique_id': unique_id,
        'forum_id': forum_id,
        'thread_id': thread_id,
        'views': views,
        'replies': replies,
        'post_id': post_id,
        'title': title,
        'unique_user_id': '', #Issue with some users not having a profile id or link
        'username': username,
        'datetime': datetime,
        'is_it_a_reply': is_it_a_reply,
        'reply_to': reply_to,
        'post_content': post_div.find('div', class_='content').get_text(strip=True),
        'post_link': post_link
    })
    unique_id += 1

pd.DataFrame(post_data)
```
